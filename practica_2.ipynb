```python
# Generamos la secuencia x(n)
n1 = np.arange(-5, 6)
x1 = rampseq(0, -5, 5)[0]
n2 = np.arange(6, 9)
x2 = np.full(n2.shape, 5)
n3 = np.arange(9, 13)
x3 = np.full(n3.shape, 10)
n = np.concatenate((n1, n2, n3))
x = np.concatenate((x1, x2, x3))

# Graficamos la secuencia x(n)
plt.figure(figsize=(10, 4))
plt.stem(n, x)
plt.title('Secuencia x(n)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()

# Generamos la secuencia x5(n)
x5 = 2 * np.roll(x, 4) + x

# Graficamos la secuencia x5(n)
plt.figure(figsize=(10, 4))
plt.stem(n, x5)
plt.title('Secuencia x5(n)')
plt.xlabel('n')
plt.ylabel('x5(n)')
plt.grid(True)
plt.show()

# Generamos la secuencia x6(n)
n6 = np.arange(-20, 21)
x6 = 0.001 * np.exp(0.5 * n6) * np.interp(n6, n, x) + 10 * np.sin(0.05 * np.pi * n6) * np.interp(n6 + 2, n, x)

# Graficamos la secuencia x6(n)
plt.figure(figsize=(10, 4))
plt.stem(n6, x6)
plt.title('Secuencia x6(n)')
plt.xlabel('n')
plt.ylabel('x6(n)')
plt.grid(True)
plt.show()

```
