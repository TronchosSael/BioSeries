a. 𝑥1(𝑛)= 3𝛿(𝑛+1)+5𝛿(𝑛+3)+3𝛿(𝑛+2)+3𝛿(𝑛)+ 𝛿(𝑛). ¿Cómo debe ser el vector de muestras?

```python
#sDefinimos n
n1, n2 = -10, 10

# Generamos y graficamos la secuencia x1(n)
x1_temp, n = impseq(-1, n1, n2)
x1 = 3 * x1_temp
x1 += 5*impseq(3, n1, n2)[0]
x1 += 3*impseq(2, n1, n2)[0]
x1 += 3*impseq(0, n1, n2)[0]
x1 += impseq(1, n1, n2)[0]

plt.figure(figsize=(12, 8))
plt.stem(n, x1, use_line_collection=True)
plt.title('Secuencia x1(n)')
plt.xlabel('n')
plt.ylabel('x1(n)')
plt.grid(True)
plt.show()

# Generamos y graficamos la secuencia x4(n)
x4_temp, n = rampseq(3, n1, n2)
x4 = 2*x4_temp
x4 -= rampseq(-2, n1, n2)[0]
x4 -= 5*stepseq(3, n1, n2)[0]

plt.figure(figsize=(12, 8))
plt.stem(n, x4, use_line_collection=True)
plt.title('Secuencia x4(n)')
plt.xlabel('n')
plt.ylabel('x4(n)')
plt.grid(True)
plt.show()
```

b. El siguiente sistema arroja una secuencia definida entre −6≤𝑛≤6


```python
# Define el nuevo rango de n
n1, n2 = -6, 6

# se genera las secuencias
u1, n = stepseq(-4, n1, n2)  # Genera u(n+4)
u2, n = stepseq(-2, n1, n2)  # Genera u(n+2)

# se calcula x2(n)
x2 = -2*u1 + 3*u2

# se grafica x2(n)
plt.figure(figsize=(8, 4))
plt.stem(n, x2, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x2(n)')
plt.title('Secuencia x2(n) = -2*u(n+4) + 3*u(n+2)')
plt.grid(True)
plt.show()
```

c. El siguiente sistema arroja una secuencia definida entre 0≤𝑛≤100

```python
# Define el rango de n
n = np.arange(0, 101)

# Calcula x3(n)
x3 = np.exp(0.01 * n) * np.sin(2 * np.pi * n)

# Grafica x3(n)
plt.figure(figsize=(8, 4))
plt.stem(n, x3, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x3(n)')
plt.title('Secuencia x3(n) = e^(0.01n) * sin(2*pi*n)')
plt.grid(True)
plt.show()
```

d. 𝑥4(𝑛)= 2𝑟(𝑛+3)−𝑟(𝑛−2)−5𝑢(𝑛−3),−10≤𝑛≤10

```python
# Define el rango de n
n1, n2 = -10, 10

# Genera las secuencias
r1, n = rampseq(-3, n1, n2, 2)  # Genera 2*r(n+3)
r2, n = rampseq(2, n1, n2)  # Genera r(n-2)
u1, n = stepseq(3, n1, n2)  # Genera u(n-3)

# Calcula x4(n)
x4 = r1 - r2 - 5*u1

# Grafica x4(n)
plt.figure(figsize=(8, 4))
plt.stem(n, x4, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x4(n)')
plt.title('Secuencia x4(n) = 2*r(n+3) - r(n-2) - 5*u(n-3)')
plt.grid(True)
plt.show()
```

```python
# Generamos la secuencia x[n]
n = np.arange(0, 18)
x = np.piecewise(n, [n <= 5, (n > 5) & (n <= 11), n > 11], [lambda n: n, lambda n: n - 5, lambda n: n - 10])

# Calculamos la derivada de x[n]
dx = np.diff(x)

# Creamos un subplot para graficar las dos secuencias
fig, axs = plt.subplots(2, figsize=(12, 8))

# Graficamos la secuencia original x[n]
axs[0].stem(n, x, use_line_collection=True)
axs[0].set_title('Secuencia x[n]')
axs[0].set_xlabel('n')
axs[0].set_ylabel('x[n]')
axs[0].grid(True)

# Graficamos la derivada de x[n]
axs[1].stem(n[:-1], dx, use_line_collection=True)  # Usamos n[:-1] porque la derivada tiene una muestra menos
axs[1].set_title('Derivada de x[n]')
axs[1].set_xlabel('n')
axs[1].set_ylabel('dx[n]')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Generamos la secuencia x(n)
n1 = np.arange(-5, 6)
x1 = rampseq(0, -5, 5)[0]
n2 = np.arange(6, 9)
x2 = np.full(n2.shape, 5)
n3 = np.arange(9, 13)
x3 = np.full(n3.shape, 10)
n = np.concatenate((n1, n2, n3))
x = np.concatenate((x1, x2, x3))

# Graficamos la secuencia x(n)
plt.figure(figsize=(10, 4))
plt.stem(n, x)
plt.title('Secuencia x(n)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()

# Generamos la secuencia x5(n)
x5 = 2 * np.roll(x, 4) + x

# Graficamos la secuencia x5(n)
plt.figure(figsize=(10, 4))
plt.stem(n, x5)
plt.title('Secuencia x5(n)')
plt.xlabel('n')
plt.ylabel('x5(n)')
plt.grid(True)
plt.show()

# Generamos la secuencia x6(n)
n6 = np.arange(-20, 21)
x6 = 0.001 * np.exp(0.5 * n6) * np.interp(n6, n, x) + 10 * np.sin(0.05 * np.pi * n6) * np.interp(n6 + 2, n, x)

# Graficamos la secuencia x6(n)
plt.figure(figsize=(10, 4))
plt.stem(n6, x6)
plt.title('Secuencia x6(n)')
plt.xlabel('n')
plt.ylabel('x6(n)')
plt.grid(True)
plt.show()

```
