Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
def stats_notas(notas):
    # Convertir el diccionario de notas a una lista
    notas_lista = list(notas.values())

    # Calcular estad√≠sticas
    minima = np.min(notas_lista)
    maxima = np.max(notas_lista)
    media = np.mean(notas_lista)
    desviacion = np.std(notas_lista)

    # Devolver las estad√≠sticas como una serie
    return {'M√≠nima': minima, 'M√°xima': maxima, 'Media': media, 'Desviaci√≥n T√≠pica': desviacion}

# Ejemplo
notas = {'Daniel': 7.3, 'Juliana': 6.9, 'Carlos': 6.4, 'Daniela': 8.2, 'David': 9.5}
print(stats_notas(notas))
```
